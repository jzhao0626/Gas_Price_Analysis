```python
import pandas as pd
import datetime
import scipy.stats as st
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
from scipy.stats import linregress
```

```python
gas_price_data = "Gas Price from 2000.csv"
gas_price_df = pd.read_csv(gas_price_data, header = 10)
new_index_df = gas_price_df.set_index('observation_date')
renamed_df = gas_price_df.rename(columns={"GASREGW":"Gas Price($)", "observation_date": "YMD"})
renamed_df
```

```python
# Generaltrends of gas price from 2000 to 2021 
new_index_df.plot.line(figsize=(20, 10))
```

```python
# Descriptive Statistics throughout the time 
renamed_df.describe()
```

```python
# See the top 10 highest price 
renamed_df.nlargest(10, ["Gas Price($)"])
```

```python
# Break down them into Year, Month, Day 
renamed_df['Year'] = pd.DatetimeIndex(renamed_df['YMD']).year
renamed_df['Month'] = pd.DatetimeIndex(renamed_df['YMD']).month
renamed_df['Day'] = pd.DatetimeIndex(renamed_df['YMD']).day
renamed_df
```

```python
renamed_df['month_year'] = pd.to_datetime(renamed_df['YMD']).dt.to_period('M')
renamed_df
```

```python
# Select the data for 2001 
first_data = renamed_df.loc[renamed_df['Year']==2001]
first_data.head()
```

```python
first_data.nlargest(10, ["Gas Price($)"])
```

```python
# Select two columns for plot
first_data_df = pd.DataFrame(first_data, columns= ['YMD','Gas Price($)'])
first_data_df = first_data_df.set_index('YMD')
first_data_df.head()
```

```python
# Draw the line graph to see the trends 
first_data_df.plot.line(figsize=(20, 10))
```

```python
# Average gas price in 2001 
first_data_df['Gas Price($)'].mean()
```

```python
# Gas price in September, 2001 
event_data = first_data.loc[first_data['Month'] ==9]
event_data.head()
```

Once we click the year, we see the trend like this above and create tooltip at the peak and show a relevant event.

```python
disasters = pd.read_csv("US Disaster List.csv")

disasters_df = disasters[["declaration_date", "incident_type"]]

disasters_df[["YMD", "Time"]] = disasters_df["declaration_date"].str.split("T", expand=True)
disasters_df[["Year","Month", "Day"]] = disasters_df["YMD"].str.split("-", expand=True)
cleaned_disasters_df = disasters_df.drop(columns = "Time")
cleaned_disasters_df
```

```python
cleaned_disasters_df["Year"] = cleaned_disasters_df["Year"].astype(int)
cleaned_disasters_df.dtypes
```

```python
filtered_disasters_df = cleaned_disasters_df.loc[cleaned_disasters_df['Year'] >= 2000]
filtered_disasters_df
```

```python
merged_df = filtered_disasters_df.merge(renamed_df, on="YMD", how="inner")
merged_df
```

```python
cleaned_df = merged_df[["YMD", "incident_type", "Gas Price($)"]]
cleaned_df = cleaned_df.drop_duplicates(keep="first")
cleaned_df
```

```python
cleaned_df.to_csv("CleanedData/Disaster_GasPrice.csv", index=False)
```
